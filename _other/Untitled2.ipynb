```python
import random
import time
import pickle
import numpy as np


file = open('../data/sequences/Despacito_200_1_1_200.txt', 'rb')
selected_points, countries_number, dictionary = pickle.load(file)
file.close()
```

```python
import matplotlib.pyplot as plt

data = []
for i in range(0, 54):
    ts = []
    for j in selected_points[0][0][:200]:
        ts.append(j[i])
    data.append(list(ts))

print(data)

for i in range(0,54):
    plt.plot(data[i])
plt.ylabel('one time series')
plt.show()

""""
d = []
for i in selected_points[0]:
    print(i)
    a = np.transpose(i)
    if(np.sum(a) >= -0.3):
        d.append(a)
        
size = len(d)
for i in range(0,54):
    plt.plot(d[0][i])
plt.ylabel('one time series')
plt.show()"""
print(dictionary)
```
